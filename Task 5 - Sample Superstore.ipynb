## Problem Statement : Perform ‘Exploratory Data Analysis’ on the provided dataset ‘SampleSuperstore’

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.set_option("display.max_columns" , None)
pd.set_option("display.max_rows" , None)
```

# #Read the SampleSuperstore file 

```python
store = pd.read_csv(r"C:\Users\home\Desktop\SPARK\SampleSuperstore.csv")
store.head()
```

```python
store.shape
```

### #Removing the country column

```python
store = store.drop(["Country"] , axis = 1)
```

```python
store.head()
```

## #Checking null counts

```python
store.isnull().sum()
```

## #Checking  value counts 

```python
store['Ship Mode'].value_counts()
```

# #Plotting 

```python
plt.figure(figsize=(18,8))
plt.bar('Sub-Category' , 'Category' , data = store)
plt.show()
```

```python
plt.figure(figsize=(10,8))

pd.crosstab(store["Quantity"],store["Category"],store["Profit"],aggfunc='sum').plot(kind="bar",stacked=True)
```

```python
plt.figure(figsize=(10,10))

store.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x="Sub-Category", hue="Region", data=store)
plt.show()
```

# #Pie- Chart----------------------

```python
plt.figure(figsize=(12,12))

store['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
plt.figure(figsize=(8,8))

sns.countplot(x="Category", hue="Region", data=store)
```

# #Scatter Plot-------------------------

```python
sns.scatterplot("Sales",'Profit',data=store)
```
